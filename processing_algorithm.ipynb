```python
import pandas as pd
df = pd.read_csv("analytics_device.csv")
df.head()
```

```python
df.columns
```

```python
ad_to_column = {}
ads = ['Ad1', 'Ad2', 'Ad3', 'Ad4', 'Ad5', 'Ad6', 'Ad7', 'Ad8']
for c in df.columns:
    ad = c[2:5]
    if ad in ads and ad in ad_to_column:
        tag = c[6:]
        ad_to_column[ad].append((tag, df[c][0]))
    elif ad in ads and not ad in ad_to_column:
        tag = c[6:]
        ad_to_column[ad] = [(tag, df[c][0])]
```

```python
print(ad_to_column)
```

```python
df = pd.DataFrame.from_dict({k: dict(v) for k, v in ad_to_column.items()}, orient='index')

# Convert all values to numeric (handling commas)
df = df.applymap(lambda x: pd.to_numeric(str(x).replace(',', ''), errors='coerce'))

# Define groups based on column names
close_cols = [col for col in df.columns if 'Close' in col]
medium_cols = [col for col in df.columns if 'Medium' in col]
far_cols = [col for col in df.columns if 'Far' in col]

# Create accumulative columns
df['Close Accumulative'] = df[close_cols].sum(axis=1)
df['Medium Accumulative'] = df[medium_cols].sum(axis=1)
df['Far Accumulative'] = df[far_cols].sum(axis=1)
df['Overall Accumulative'] = df.sum(axis=1)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df = df.sort_index()
```
